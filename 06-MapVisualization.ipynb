## Folium-choropleth를 이용한 미국 실업률 시각화

```python
import pandas as pd

state_data = pd.read_csv('./data/03_US_Unemployment_Oct2012.csv')
state_data.head()
```

```python
state_data
```

```python
import json

m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data='./data/03_us-states.json',
    data=state_data, columns=['State', 'Unemployment'],
    nan_fill_color='purple', nan_fill_opacity=0.4,
    key_on='feature.id', fill_color='YlGn',
    legend_name='Unemployment Rate (%)'
).add_to(m)

m
```

## JSON은 알고 가자~

```python
import json

customer = {
    'id': '000001',
    'name': '홍길동',
    'history': [
        {'date': '2018-05-22', 'log': True},
        {'date': '2018-05-23', 'log': False},
    ]
}
```

```python
# JSON 인코딩
json_test = json.dumps(customer, indent=4, ensure_ascii=False)

# 문자열 출력
print(json_test)
```

```python
# Write JSON
with open('./data/03_json_tutorial.json', 'w', encoding='utf=8') as make_file:
    json.dump(customer, make_file, ensure_ascii=False, indent=4)
```

```python
with open('./data/03_us-states.json') as f:
    json_test = json.load(f)
```
